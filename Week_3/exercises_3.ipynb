```python
# Exercise 1
class Item:
    def __init__(self, name, price_history):
        self.name = name
        self.price_history = price_history
    
    def __repr__(self):
        return self.name
    
desk = Item("Desk", [75, 100, 125, 150])
chair= Item("Chair", [25, 35, 45, 55])
bed_frame = Item("Bed frame", [125, 150, 175, 200])
mattress = Item("Mattress", [125, 150, 175, 200])
carpet = Item("Carpet", [30, 40, 50, 60])

class Store:
    def __init__(self, name, city, coordinates, items):
        self.name = name
        self.city = city
        self.coordinates = coordinates
        self.items = items
    
    def __repr__(self):
        return f"{self.name} is located in {self.city}"

BB = Store("Beds and Beyond", "Leiden", [123, 321], {bed_frame, mattress})
NJD = Store("Not Just Desks", "Leiden", [456, 654], {desk, chair})
SA = Store("Soft Accoutrements", "Oegstgeeest", [789, 987], {carpet, mattress, chair})
print()

# Exercise 2
only_Leiden = (BB.items | NJD.items)
print(only_Leiden)

Leiden_and_Oegstgeest = only_Leiden & SA.items
print(Leiden_and_Oegstgeest)

only_Oegstgeest = SA.items
print(only_Oegstgeest, "\n")

# Exercise 3
store_locations = {
    tuple(BB.coordinates) : BB,
    tuple(NJD.coordinates) : NJD,
    tuple(SA.coordinates) : SA
}

for coordinates, store in store_locations.items():
    print(f"{coordinates}: {store.name}")
print()

# Exercise 4
for (longtitude, latitude), store in store_locations.items():
    print(f"{store.name} ({store.city}) is found at longtitude {longtitude}, and latitude {latitude}.")

for item in only_Leiden:
    first_price, *_, last_price = item.price_history
    difference = last_price - first_price
    print(f"the difference in price for a {item} is {difference}")
```

```python
sentence = "jnrfskvd mbcngreujnkdfvjnsfngiognrgneog"

def count_letters(text):
    counts = {}
    for ch in text.lower().strip():
        if ch.isalpha():
            counts[ch] = counts.get(ch,0) + 1  # (get character value, default if non-existent)
    return counts



def count_letter_position(text):
    positions = {}
    text = text.lower()

    for index, ch in enumerate(text, start=1): 
        if ch.isalpha():
            if index not in positions:
                positions[index] = {}  # New key-value pair with index as key and dict as value
            positions[index][ch] = positions[index].get(ch, 0) + 1  # New key-value pair inside the just made dict with ch as key and new value as value
    return positions

count_letters(sentence)
count_letter_position(sentence)
```

```python
def unique(text):
    unique_chars = set()

    for ch in text.lower():
        if ch.isalpha():
            unique_chars.add(ch)  # add for sets and append for lists
    return unique_chars, len(unique_chars)

def unique(text):
    unique_chars = []

    for ch in text.lower():
        if ch.isalpha():
            unique_chars.append(ch)  # add for sets and append for lists
    return unique_chars, len(unique_chars)
unique("hgvjbknlm;,")
```

```python
some_variable = -3
new_list = [some_variable] * 4 # All 4 elements point to the same MOV
new_list[0] = 1  # Constants are immultabe so the same MOV doesn't matter
print(new_list)
```

```python
some_variable = [-3]
new_list = [some_variable] * 4  # All 4 elements point to the same MOV
new_list[0][0] = 1  # Lists are mutable so you change all MOV
print(new_list)
```

```python
def sequence_replicator(element, length):
    new_list = []
    for i in range(length):
        new_list.append(element.copy())   # Copy creates a new MOV`
    return new_list

some_variable = [-3]
new_list = sequence_replicator(some_variable, 4)
new_list[0][0] = 1 

new_list
```

```python
def search(list, target):
    for i, value, in enumerate(list):
        if value == target:
            index = i
            break
    return index

list = [1,2,3,4,5,6]

search(list, 4)
```
