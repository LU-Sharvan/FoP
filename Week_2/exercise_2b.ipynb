```python
word = "kip"
n_lives = 5
n_known = 0
end = 0
known =["_"] * len(word) #makes a list ["_", "_", etc]

print("Welcome to the hangman game!")
print("----------------------------")

while n_lives > end and len(word) != n_known:
    print("Known letters:", "".join(known)) #makes list a continuous line
    print(f"Lives left: {n_lives}")
    letter = input("Enter letter")
    if len(letter) != 1:
        print("----------------------------")
        continue
    
    new_found_letters = 0 #resets every loop
    for x, y in enumerate(word): #y = letter of word at index x
        if known[x] == "_" and y == letter:
            known[x] = letter
            new_found_letters = 1
    
    if new_found_letters > 0: #always 0 or 1
        n_known += new_found_letters
    else:
        n_lives -= 1
print("-----------------------------")

if n_known == len(word):
    print(f"You won! The word was {word}")
else:
    print("Sorry, you ran out of lives")
```
