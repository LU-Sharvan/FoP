```python
#Initializes variables
word = "aloo"
n_lives = 5
n_known = 0
end = 0
known =["_"] * len(word) #makes a list ["_", "_", etc]

#Welcomes players to the game
print("Welcome to the hangman game!")
print("----------------------------")

#repeated message in the game
while n_lives > end and "_" in known: #checks if you're not dead and if not every character has been guessed
    print("Known letters:", " ".join(known)) #makes list a continuous line
    print(f"Lives left: {n_lives}")
    letter = input("Enter letter")
    print("----------------------------")

    #no life is lost when entering 2+ symbols
    if len(letter) != 1 or not letter.isalpha():
        print("----------------------------")
        continue

   #resets count every loop
    new_found_letters = 0
    
    #replaces "_" in known with a letter y
    for x, y in enumerate(word): #y is letter of secret word at index x
        if known[x] == "_" and y == letter:
            known[x] = letter #replaces "_" with right letter
            new_found_letters = 1 #1 for every combination of x and y
    
    #either increases the amount of known letters or decreases the amount of lives
    if new_found_letters > 0: 
        n_known += new_found_letters
    else:
        n_lives -= 1

#determines win or lose
if "_" not in known: #if everything has been guessed
    print("Known letters:", " ".join(known))
    print(f"Lives left: {n_lives}")
    print(f"You won! The word was {word}")
else: #if not everything has been guessed
    print("Sorry, you ran out of lives")
```

```python
word = "balloon"
known = ["_"] * len(word)
guess = "l"
newly_found = 0

for i, ch in enumerate(word):
    if ch == guess and known[i] == "_":
        known[i] = guess
        newly_found += 1

print("known:", "".join(known))
print("newly_found:", newly_found)
```
