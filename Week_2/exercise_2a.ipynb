```python
fibonacci = [0,1,1,2,3,5,8,13,21]
print(len(fibonacci))
print(fibonacci[-2:])
fibonacci.append(fibonacci[-1] + fibonacci[-2])
print(len(fibonacci))
print(max(fibonacci))
print(sum(fibonacci[-2:]))
eleventh_element = [55]
print(fibonacci + eleventh_element)
fibonacci.remove(0)
print(fibonacci)
print(fibonacci[::-1])
fibonacci.reverse()
print(fibonacci)
```
